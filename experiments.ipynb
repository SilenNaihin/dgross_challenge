```python
# https://chat.openai.com/share/67f500ad-b70a-4fe5-b287-8b6af9a31d73

from scipy.signal import spectrogram
import matplotlib.pyplot as plt
import numpy as np
from scipy.io import wavfile

# Step 1: Extract Audio from Video using ffmpeg-python
input_video = './oEA8MDQU0B4bX45r.mp4' # Path to your video file
output_audio = './audio.wav' # Output path for the extracted audio

# Step 2: Load the extracted audio file
sample_rate, samples = wavfile.read(output_audio)

# Check if the audio is stereo and select the first channel if so
if samples.ndim == 2:
    samples = samples[:, 0]  # Select the first channel for mono

# Step 3: Generate Spectrogram
frequencies, times, spectrogram_data = spectrogram(samples, sample_rate)

# Step 4: Plot the Spectrogram
plt.figure(figsize=(20, 10))
plt.pcolormesh(times, frequencies, 10 * np.log10(spectrogram_data), shading='gouraud')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.title('Spectrogram')
plt.colorbar(label='Intensity [dB]')
plt.show()
```
